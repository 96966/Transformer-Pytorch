```python
class Solution(object):
    def minWindow(self, s, t):
        """
        :type s: str
        :type t: str
        :rtype: str
        """
        # 构建字典：
        t_dict = {}
        for char in t:
            t_dict[char] = t_dict.get(char, 0) + 1
        
        min_len = float('inf')
        # 判断是否匹配： valid == n
        valid = 0
        # 左右指针，左边先不动，右边动
        left, right = 0, 0
        # 由左右指针构建的窗口
        window = {}
        while right < len(s):
            r = s[right]
            right +=1

            # 先把r存进window中
            window[r] = window.get(r,0) + 1

            # 然后判断r是不是在t中
            if r in t_dict:
                # 再比对r字符是否已经满足
                if window[r] == t_dict[r]:
                    valid +=1
            
            while valid == len(t_dict):
                min_len = min(min_len, right - left + 1)

                l = s[left]
                if l in t_dict:
                    if window[l] == t_dict[l]:
                        valid -=1
                        window[l] -=1
                left +=1
        return s[left-1:right] if min_len < float('inf') else ""
            
```
